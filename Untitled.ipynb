# Data cleaning

```python
data.drop_duplicates(inplace=True)
```

```python
sns.distplot(data['bmi'])

plt.show()
```

```python
data['bmi'].skew()
```

```python
# z-score
upper_limit=data['bmi'].mean()+3*data['bmi'].std()
lower_limit=data['bmi'].mean()-3*data['bmi'].std()

# capping
data['bmi']=np.where(
    data['bmi']>upper_limit,
    upper_limit,
    np.where(
        data['bmi']<lower_limit,
        lower_limit,
        data['bmi']
    )
    
)
```
